## 1. Consolidación de inventarios 
Combina tres inventarios provenientes de distintas sucursales. Cuando un producto aparezca en varias 
sucursales, suma sus cantidades y calcula también en cuántas sucursales está disponible. 

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3} 
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5} 
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7} 
Requisitos:
 
- Crea un diccionario final con las claves cantidad_total y sucursales. 
- No modifiques los diccionarios originales. 
- Ordena el resultado alfabéticamente por producto.


```python
inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3} 
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5} 
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}

inventario_total={}
for llave,valor in inventario_norte.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
    inventario_total[llave]+=valor
for llave,valor in inventario_sur.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
    inventario_total[llave]+=valor
for llave,valor in inventario_centro.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
    inventario_total[llave]+=valor

print(inventario_total)
llaves_ordeenadas=sorted(inventario_total.keys())
inventario_con_llaves_ordenadas={}
for llave in llaves_ordeenadas:
    inventario_con_llaves_ordenadas[llave]=inventario_total[llave]
print(inventario_con_llaves_ordenadas)
```

## 2. Ranking de estudiantes con desempate 
Calcula el promedio de cada estudiante y genera un ranking de mayor a menor. Si dos estudiantes tienen el 
mismo promedio, ordénalos alfabéticamente. 
calificaciones = { 
"Ana": [17, 18, 16, 19], 
"Luis": [15, 16, 17, 16], 
"Carla": [18, 17, 19, 18], 
"Pedro": [16, 16, 16, 16], 
"Lucía": [15, 17, 16, 16] 
} 
Requisitos:
 
- Guarda promedio, nota_máxima y nota_mínima por estudiante. 
- El ranking final debe ser una lista de tuplas.

```python
calificaciones = { 
"Ana": [17, 18, 16, 19], 
"Luis": [15, 16, 17, 16], 
"Carla": [18, 17, 19, 18], 
"Pedro": [16, 16, 16, 16], 
"Lucía": [15, 17, 16, 16] 
} 
estadisticas = {}

for estudiante, notas in calificaciones.items():
    print(estudiante, notas)
    estadisticas[estudiante] = {
        "minima_nota": min(notas),
        "maxima_nota": max(notas),
        "promedio": sum(notas) / len(notas)
    }
    print(estadisticas[estudiante])
    print('='*20) 
```

```python
for estudiante, stats in estadisticas.items():
    print(estudiante)
    print(stats)
    print('='*20)
```

```python
ranking = []
for nombre, stats in estadisticas.items():
    ranking.append(
        (nombre, stats["promedio"])
    )
print(ranking)

# ranking = sorted(ranking, key=lambda x: x[1], reverse=True)
# print(ranking)
```

```python
ranking = sorted(ranking, key=lambda item: item[1], reverse=True)
print(ranking)
```

Agrupa empleados primero por departamento y luego por nivel de experiencia. 
empleados = [ 
{"nombre": "Ana", "departamento": "Desarrollo", "nivel": "Senior", "salario": 
5200}, 
{"nombre": "Luis", "departamento": "Diseño", "nivel": "Junior", "salario": 2800}, 
{"nombre": "Carla", "departamento": "Desarrollo", "nivel": "Junior", "salario":

{"nombre": "Pedro", "departamento": "Desarrollo", "nivel": "Senior", "salario": 
5600}, 
{"nombre": "Lucía", "departamento": "Diseño", "nivel": "Senior", "salario": 4700} 
] 
Requisitos:
 
- En cada grupo almacena nombres, cantidad y salario promedio. 
- La estructura debe ser un diccionario anidado. 

```python
empleados = [
    {"nombre": "Ana", "departamento": "Desarrollo", "nivel": "Senior", "salario": 5200},
    {"nombre": "Luis", "departamento": "Diseño", "nivel": "Junior", "salario": 2800},
    {"nombre": "Carla", "departamento": "Desarrollo", "nivel": "Junior", "salario": 3400},
    {"nombre": "Pedro", "departamento": "Desarrollo", "nivel": "Senior", "salario": 5600},
    {"nombre": "Lucía", "departamento": "Diseño", "nivel": "Senior", "salario": 4700}
]
```

```python
resumen = {}

for idx, empleado in enumerate(empleados):
    # print(empleado)
    departamento = empleado['departamento']
    nivel = empleado['nivel']
    nombre = empleado['nombre']
    salario = empleado['salario']

    if departamento not in resumen:
        resumen[departamento] = {}

    if nivel not in resumen[departamento]:
        resumen[departamento][nivel] = {
            "cantidad": 0,
            "nombres": [],
            "salarios": []
        }

    grupo = resumen[departamento][nivel]
    grupo["nombres"].append(nombre)
    grupo["salarios"].append(salario)
    grupo["cantidad"] += 1


for registro, grupo_xp in resumen.items():
    for experiencia, valor in grupo_xp.items():
        valor["promedio"] = sum(valor["salarios"]) / len(valor['salarios'])
print(resumen)
```

## 7. Sistema de votación con validación 
Procesa una lista de votos y separa los votos válidos de los inválidos. 
candidatos = {"Ana", "Luis", "Carla"} 
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"] 
Requisitos:
 
- Los votos deben ser insensibles a mayúsculas. 
- Registra conteo por candidato, votos inválidos y porcentaje obtenido. 
- Determina si hay ganador o empate.

```python
candidatos = {"Ana", "Luis", "Carla"}
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"]

mapa = {
    "ana": "Ana",
    "luis": "Luis",
    "carla": "Carla"
}

conteo = {}

for voto in votos:
    voto_limpio = voto.lower()   # "ana"
    candidato = mapa.get(voto_limpio, "invalido")

    if candidato == "invalido":
        conteo.setdefault("Votos invalidos", 0)
        conteo["Votos invalidos"] += 1
        continue

    conteo.setdefault(candidato, 0) # {"Ana": 0}
    conteo[candidato] += 1  # conteo["Ana"] = 0 + 1 -> conteo["Ana"] = 1


total_votos = len(votos)
for candidato in candidatos:
    conteo[f"{candidato}_pct"] = round(conteo[candidato] / total_votos * 100, 2)

conteo["Votos invalidos pct"] = round(conteo["Votos invalidos"] / total_votos * 100, 2)
```

```python
print(conteo)
```
