## 1 Analizar las ventas de una tienda
 
Enunciado:
 Crea una función llamada analizar_ventas(ventas) que reciba una lista de montos de venta y 
construya un resumen estadístico. 
Requisitos:
- Calcular el total vendido, el promedio, la venta más alta y la venta más baja. 
- Contar cuántas ventas fueron superiores al promedio. 
- Indicar cuántas ventas se encuentran por debajo de 100. 
- Retornar todos los resultados dentro de un diccionario. 
- Si la lista está vacía o contiene valores negativos, retornar un mensaje de error. 

Datos de prueba: 
ventas = [150.50, 200, 80.90, 300, 175.50, 90] 
resultado = analizar_ventas(ventas) 
Reto
 
adicional:
 Agrega un parámetro opcional meta. La función debe indicar si el total vendido alcanzó dicha 
meta. 

```python
ventas_1 = [150.50, 200, 80.90, 300, 175.50, 90]
```

```python
def analizar_ventas(ventas:list[int|float])->dict[str,int|float]:
    if not ventas:
        return " no hay ventas"
    for venta in ventas:
        if venta<0:
            return "hay  ventas"
    total_vendido=sum(ventas)
    promedio=total_vendido/len(ventas)
    venta_mas_alta=max(ventas)
    venta_mas_baja=min(ventas)
    total_superior_vendido=0
    total_venta_inferior=0
    contar=0
    for venta in ventas: 
        if venta > promedio:
            total_superior_vendido=total_superior_vendido+1
        if venta < promedio:
            total_venta_inferior=total_venta_inferior+1
    return{
        "total vendido":total_vendido,
        "promedio":promedio,
        "venta mas baja":venta_mas_baja,
        "Venta mas alta":venta_mas_alta,
        "total de ventas altas":total_superior_vendido,
        "total de ventas bajas":total_venta_inferior}

analizar_ventas(ventas_1)
```
