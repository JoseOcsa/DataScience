## 2. Validar una contraseña y explicar los errores
Crea una función validar_password(password) que determine si una contraseña cumple todas 
las reglas de seguridad. 
Requisitos:
 
- Debe tener al menos 8 caracteres. 
- Debe contener una letra mayúscula, una minúscula y un número. 
- Debe incluir al menos uno de estos símbolos: @, #, $ o %. 
- No debe contener espacios. 
- Debe retornar una tupla con el formato (es_valida, errores), donde errores sea una lista.

```python
# Ejercicio 2
def validar_password(password: str) -> tuple[bool, list[str]]:

    simbolos_requeridos = ("@", "#", "$",  "%")

    errores = []

    if len(password) < 8:
        errores.append(
            "El password tiene menos de 8 caracteres"
        )

    contiene_minuscula = False
    contiene_mayuscula = False
    contiene_numero = False
    contiene_simbolos_requeridos = False

    for caracter in password:
        if caracter.islower():
            contiene_minuscula = True
            break

    for caracter in password:
        if caracter.isupper():
            contiene_mayuscula = True
            break

    for caracter in password:
        if caracter.isnumeric():
            contiene_numero = True
            break

    for caracter in password:
        if caracter in simbolos_requeridos:
            contiene_simbolos_requeridos = True
            break

    if not contiene_minuscula:
        errores.append(
            "El password no contiene letra minuscula"
        )

    if not contiene_mayuscula:
        errores.append(
            "El password no contiene letra mayuscula"
        )

    if not contiene_numero:
        errores.append(
            "El password no contiene caracter numerico"
        )

    if not contiene_simbolos_requeridos:
        errores.append(
            "El password no contiene simbolos requeridos"
        )

    if len(password) != len(password.strip()):
        errores.append(
            "El password contiene espacios en blanco"
        )

    es_valida = False if errores else True

    return (es_valida, errores)
```

```python
es_valida, errores = validar_password('python@')
print(es_valida)
print(errores)
```

```python
def registrar_estudiante(
    estudiantes: dict[str, int | float | str],
    nombre: str,
    notas: list[int | float]
) -> str:
    if not nombre:
        return "Nombre invalido"

    if nombre in estudiantes:
        return "El nombre ya existe, no se admite duplicados"

    for nota in notas:
        if isinstance(nota, str):
            return "Hay notas no numericas"
        else:
            if nota < 0 or nota > 20:
                return "Hay notas con valores menores a 0 o mayores a 20"

    promedio = round(sum(notas) / len(notas), 2)

    if promedio >= 11:
        estado = "Aprobado"
    else:
        estado = "Desaprobado"

    estudiantes[nombre] = {
        "estado": estado,
        "promedio": promedio,
        "notas": notas
    }

    return "El registro fue exitoso"
```

```python
estudiantes = {}

registrar_estudiante(
    estudiantes,
    'Ana',
    [18, 15, 17]
)
```

## 4. Agrupar palabras por longitud
Crea una función agrupar_por_longitud(palabras) que organice una lista de palabras dentro de 
un diccionario. 
Requisitos:
 
- La clave debe ser la cantidad de caracteres de cada palabra. 
- El valor debe ser una lista con todas las palabras de esa longitud. 
- Ignorar elementos que no sean cadenas de texto. 
- Eliminar palabras duplicadas sin importar mayúsculas y minúsculas. 
- Ordenar alfabéticamente las palabras de cada grupo. 


```python
def agrupar_por_longitud(texto:list|str)->list|str:
    minuscula=[]
    sin_repetir=[]
    resultado={}
    #Primero convertir texto a minuscula
    for palabra in texto:
        minuscula.append(palabra.lower())
    for palabra in minuscula:
        if palabra not in sin_repetir:
            sin_repetir.append(palabra)
    sin_repetir.sort()
    for palabra in sin_repetir:
        longitud = len(palabra)
        if longitud not in resultado:
            resultado[longitud]=[]
        
        resultado[longitud].append(palabra)
        
    return resultado
```

```python
palabras = [ 'python', 'Java', 'código', 'web', 'java', 'datos', 'API', 'lista' ] 
agrupar_por_longitud(palabras)
```
