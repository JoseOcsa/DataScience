## 2. Validar una contraseña y explicar los errores
Crea una función validar_password(password) que determine si una contraseña cumple todas 
las reglas de seguridad. 
Requisitos:
 
- Debe tener al menos 8 caracteres. 
- Debe contener una letra mayúscula, una minúscula y un número. 
- Debe incluir al menos uno de estos símbolos: @, #, $ o %. 
- No debe contener espacios. 
- Debe retornar una tupla con el formato (es_valida, errores), donde errores sea una lista.

```python

def validar_password(password: str) -> tuple[bool, list[str]]:

    simbolos_requeridos = ("@", "#", "$",  "%")

    errores = []

    if len(password) < 8:
        errores.append(
            "El password tiene menos de 8 caracteres"
        )

    contiene_minuscula = False
    contiene_mayuscula = False
    contiene_numero = False
    contiene_simbolos_requeridos = False

    for caracter in password:
        if caracter.islower():
            contiene_minuscula = True
            break

    for caracter in password:
        if caracter.isupper():
            contiene_mayuscula = True
            break

    for caracter in password:
        if caracter.isnumeric():
            contiene_numero = True
            break

    for caracter in password:
        if caracter in simbolos_requeridos:
            contiene_simbolos_requeridos = True
            break

    if not contiene_minuscula:
        errores.append(
            "El password no contiene letra minuscula"
        )

    if not contiene_mayuscula:
        errores.append(
            "El password no contiene letra mayuscula"
        )

    if not contiene_numero:
        errores.append(
            "El password no contiene caracter numerico"
        )

    if not contiene_simbolos_requeridos:
        errores.append(
            "El password no contiene simbolos requeridos"
        )

    if len(password) != len(password.strip()):
        errores.append(
            "El password contiene espacios en blanco"
        )

    es_valida = False if errores else True

    return (es_valida, errores)
```

## 3. Registrar estudiantes con validaciones
 Crea una función registrar_estudiante(estudiantes, nombre, notas) que agregue la información 
de un estudiante a un diccionario. 
Requisitos:
 
- No permitir nombres vacíos ni nombres duplicados. 
- Cada estudiante debe tener como mínimo tres notas. 
- Todas las notas deben ser numéricas y estar entre 0 y 20. 
- Calcular el promedio y guardarlo con dos decimales. 
- Guardar el estado Aprobado si el promedio es mayor o igual a 11; de lo contrario, Desaprobado. 
- Retornar un mensaje que indique si el registro fue exitoso

```python
def registrar_estudiante(
    estudiantes: dict[str, int | float | str],
    nombre: str,
    notas: list[int | float]
) -> str:
    if not nombre:
        return "Nombre invalido"

    if nombre in estudiantes:
        return "El nombre ya existe, no se admite duplicados"

    for nota in notas:
        if isinstance(nota, str):
            return "Hay notas no numericas"
        else:
            if nota < 0 or nota > 20:
                return "Hay notas con valores menores a 0 o mayores a 20"

    promedio = round(sum(notas) / len(notas), 2)

    if promedio >= 11:
        estado = "Aprobado"
    else:
        estado = "Desaprobado"

    estudiantes[nombre] = {
        "estado": estado,
        "promedio": promedio,
        "notas": notas
    }

    return "El registro fue exitoso"
```

## 4. Agrupar palabras por longitud
Crea una función agrupar_por_longitud(palabras) que organice una lista de palabras dentro de 
un diccionario. 
Requisitos:
 
- La clave debe ser la cantidad de caracteres de cada palabra. 
- El valor debe ser una lista con todas las palabras de esa longitud. 
- Ignorar elementos que no sean cadenas de texto. 
- Eliminar palabras duplicadas sin importar mayúsculas y minúsculas. 
- Ordenar alfabéticamente las palabras de cada grupo. 


```python
# FORMA 1
def agrupar_por_longitud(texto:list|str)->list|str:
    minuscula=[]
    sin_repetir=[]
    resultado={}
    #Primero convertir texto a minuscula
    for palabra in texto:
        minuscula.append(palabra.lower())
    for palabra in minuscula:
        if palabra not in sin_repetir:
            sin_repetir.append(palabra)
    sin_repetir.sort()
    for palabra in sin_repetir:
        longitud = len(palabra)
        if longitud not in resultado:
            resultado[longitud]=[]
        
        resultado[longitud].append(palabra)
        
    return resultado
```

```python
palabras = [ 'python', 'Java', 'código', 'web', 'java', 'datos', 'API', 'lista' ] 
agrupar_por_longitud(palabras)
```

```python
# FORMA 2 (hecho por el profesor)
def agrupar_por_longitud1(palabras:list[str])->dict[int,list[str]]:
    #Devolver palabras unicas
    palabras_limpias=set([
        palabra.lower()
        for palabra in palabras
        if isinstance(palabra,str)
    ])
    grupos={}
    for palabra_limpia in palabras_limpias:
        longitud1=len(palabra_limpia)
        grupos.setdefault(longitud1,[]).append(palabra_limpia)

    for llave in grupos:
        grupos[llave].sort()
    
    return grupos
```

- La operación solo puede ser entrada o salida. 
- La cantidad debe ser un entero positivo. 
- Una entrada puede crear un producto nuevo o aumentar su stock. 
- Una salida solo puede realizarse si el producto existe y tiene suficiente stock.

## 6. Calcular el costo completo de un pedido
Crea una función calcular_pedido(productos, cantidades, descuento=0, delivery=0) que genere 
el resumen de una compra. 
Requisitos:
 
- productos será un diccionario con el nombre y el precio de cada producto. 
- cantidades será un diccionario con los productos solicitados y sus cantidades. 
- Ignorar productos inexistentes, pero incluirlos en una lista de no_encontrados. 
- Validar que las cantidades sean enteros positivos. 
- Calcular subtotal, descuento, delivery y total final. 
- El descuento debe encontrarse entre 0 y 100. 
- Retornar todos los resultados en un diccionario. 

```python
def calcular_pedido(
        Productos: int|str,
        Cantidades: int|str,
        Descuentos:int|float,
        Delivery:int|float):
    
    total=0
    no_encontrados=[]
    #Validar si el descuento es viable
    if Descuentos<0 or Descuentos>100:
        return {"Error":"Descuento Invalido"}
    for producto,cantidad in Cantidades.items():

        # Validar si es entero las cantidades propuestas,sino manda error
        if type(cantidad) != int or cantidad <=0:
            return {"Error":"Dato invalido"}
        #Verificar existencia de los productos
        if producto in Productos:
            total=total+Productos[producto]*cantidad
        else:
            no_encontrados.append(producto)
    #Calculos para el total de precio *cantidad
    monto_des=total*Descuentos/100
    monto_total=monto_des+Delivery

    return{
        "Sub Total":total,
        "Descuento":monto_des,
        "Monto Final":monto_total,
        "delivery":Delivery,
        "Productos no encontrados":no_encontrados    }
```

```python
productos = {'Hamburguesa': 15, 'Pizza': 25, 'Gaseosa': 5} 
cantidades = {'Hamburguesa': 2, 'Gaseosa': 3, 'Helado': 1} 
calcular_pedido(productos,cantidades,25,20.5)
```

## 7. Evaluar el rendimiento de empleados
Crea una función evaluar_empleados(empleados, minimo=70) que analice las evaluaciones de 
cada empleado. 
Requisitos:
 
- Cada empleado tendrá una lista de puntajes entre 0 y 100. 
- Calcular el promedio de cada empleado. 
- Clasificar el resultado como Excelente, Bueno, Regular o Deficiente. 
- Incluir únicamente a quienes tengan un promedio igual o superior a minimo. 
- Retornar una lista de tuplas con (nombre, promedio, clasificación). 
- Ordenar de mayor a menor promedio y, en caso de empate, alfabéticamente. 


```python
empleados = { 
'Ana': [80, 90, 75], 
'Carlos': [60, 70, 65], 
'Lucía': [95, 85, 90], 
'Pedro': [70, 70, 70]}
```

```python
def evaluar_empleados(empleados:dict[str|list],minimo:int):
    resultado=[]
    for name,evaluaciones in empleados.items():
        prom=sum(evaluaciones)/len(evaluaciones)
        if prom >= 90:
            calificacion="Excelente"
        elif prom >=70:
            calificacion="Bueno"
        elif  prom >= 50:
            calificacion="Regular"
        else:
            calificacion="Deficiente"

        if prom >=minimo:
            resultado.append((name,prom,calificacion))
    resultado.sort(key=lambda x:x[1], reverse=True)
        
    return resultado
evaluar_empleados(empleados,50)
```

## 8. Comprimir y  descomprimir texto
 Implementa dos funciones: comprimir_texto(texto) y descomprimir_texto(texto). 
Requisitos:
 
- comprimir_texto debe reemplazar caracteres consecutivos repetidos por el carácter y su cantidad. 
- Por ejemplo, aaabbccccdaa debe convertirse en a3b2c4d1a2. 
- descomprimir_texto debe reconstruir el texto original. 
- La descompresión debe admitir cantidades de más de un dígito, como a12. 
- Si el formato comprimido es inválido, retornar un mensaje de error.
 

```python


def comprimir_texto(texto:str) -> str:
    resultado=[]
    caracter_actual=texto[0]
    contador =1
    for caracter in texto[1:]:
        if caracter == caracter_actual:
            contador += 1
        else:
            resultado.append(f"{caracter_actual}{contador}")
            caracter_actual=caracter
            contador =1
    resultado.append(f"{caracter_actual}{contador}")
    #Concatenador en la lista mediante "".join
    resultado="".join(resultado)
    return resultado

texto ="aaabbccccdaa"
comprimir_texto(texto)
```

```python
def descomprimir_texto(texto:str)->str:
    letras=texto[0::2]
    numeros=texto[1::2]
    resultado=[]
    for letra,numero in zip(letras,numeros):
        resultado.append(letra*int(numero))
    resultado="".join(resultado)
    return resultado

texto_incriptado="a3b2c4d1a2"
descomprimir_texto(texto_incriptado)
```

## 10.  Sistema de reservas de asientos
 Crea una función reservar_asientos(sala, asientos_solicitados) que permita reservar varios 
asientos de una sala. 
Requisitos:
 
- La sala será una lista de listas donde L significa libre y O significa ocupado. 
- Cada asiento solicitado se expresará mediante una tupla (fila, columna). 
- Validar que todas las posiciones existan y que los asientos estén libres. 
- Realizar la reserva únicamente si todos están disponibles. 
- Evitar reservas parciales y mantener la sala intacta si ocurre un error. 
- Retornar el estado, los asientos reservados y la sala actualizada. 

```python
def reservar_asientos(sala, asientos_solicitados):

    reservados = []

    # Verificar todos los asientos
    for fila, columna in asientos_solicitados:

        # Verificar que exista la fila
        if fila < 1 or fila > len(sala):
            return {"Resultado": "Fila no existe"}

        # Verificar que exista la columna
        if columna < 1 or columna > len(sala[fila-1]):
            return {"Resultado": "Columna no existe"}

        # Verificar que el asiento esté libre
        if sala[fila-1][columna-1] != "L":
            return {"Resultado": "Ocupado"}

    # Si todos están libres, realizar la reserva
    for fila, columna in asientos_solicitados:
        sala[fila-1][columna-1] = "O"
        reservados.append((fila, columna))

    return {
        "Resultado": "Reserva realizada",
        "Sala": sala
    }
```
