## 1. Buscar el mayor y el menor valor 
Dada la tupla numeros = (15, 8, 42, 3, 27, 19, 5), obtén el número mayor, el menor y la 
diferencia entre ambos. 

```python

tupla = (15, 8, 42, 3, 27, 19, 5)
minimo_numero = min(tupla)
maximo_numero = max(tupla)
diff = maximo_numero - minimo_numero
print(
    f"Maximo numero: {maximo_numero}\nMinimo numero: {minimo_numero}\ndiff: {diff}"
)
```


## 2. Contar ocurrencias 
Dada la tupla datos = (5, 8, 5, 1, 9, 5, 2, 8, 5, 3), solicita un número al usuario y muestra 
cuántas veces aparece sin convertir la tupla en lista.

```python
datos = (5,8, 5, 1, 9, 5, 2, 8, 5, 3)
Entrada=int(input("Ingresa numero que deseas buscar:"))
contado=datos.count(Entrada)
print(f"Las veces que aparece el {Entrada} son {contado} veces")
```

## 3. Encontrar la primera posición 
Dada la tupla nombres = ('Ana','Luis','Pedro','María','Luis','Carlos'), solicita un nombre y 
muestra su primera posición o un mensaje si no existe.

```python
nombres = ('Ana','Luis','Pedro','María','Luis','Carlos')
name=str(input("Ingresa el nombre que desees:"))

for nombre in nombres:
    if name == nombre:
        tupla1=nombres.index(name)
        print(f"El nombre de {name} se encuentra en la posicion {tupla1}")
if name not in nombres:
    print(f"El nombre de {name} no existe")
    
```

## 4. Estadísticas de una tupla 
Con notas = (18,15,20,11,14,17,19), calcula promedio, máxima, mínima y cantidad de 
aprobados. 

```python
notas = (18,15,20,11,14,17,19)
minimo = min(notas)
maximo = max(notas)
prom = round(sum(notas)/len(notas),2)
aprobados=0
print(
    f"Maxima nota: {maximo}\nMinima nota: {minimo}\nPromedio: {prom}"
)
for nota in notas:
    if nota> prom:
        aprobados += 1
print(F"La cantidad de aprobados son: {aprobados} estudiantes")
        
```
