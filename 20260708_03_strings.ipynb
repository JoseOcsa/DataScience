## FUNCIONES  u operadores en  strings
-longitud de mi cadena(len)
-convertir todo a minuscula(lower)
-convertir todo a mayuscula (upper)
-Reemplazar un valor de la cadena (replace)
-"    texto " strip -> (texto)
-texto1,texto2,texto3 --> (split,",")>> [texto1 ,texto2,texto3]

```python
nombre="Jose"
print("cuantas letras tiene mi nombre:",len(nombre))
```

```python
print(nombre.upper())
```

```python
print(nombre.lower())
```

```python
# Replace acepta dos paramentros (lo que quiero reemplazar, lo que voy a reemplazar)
print(
    nombre.replace("e","lo")
)
```

```python
nombre= "Jose                        "
print(len(nombre))

nombre_limpio=nombre.strip()
print(len(nombre_limpio))
print(nombre_limpio)

nombre= "xxJosex" 
print(len(nombre))
nombre_limpio=nombre.strip("x")
print(len(nombre_limpio))
print(nombre_limpio)
```

```python
nombres1= "Jose Ocsa"
nombres_separados1=nombres1.split(" ")
print(nombres_separados1)
nombres= "JosexxOcsa"
nombres_separados=nombres.split("xx")
print(nombres_separados)
```

```python
nombre="Jose"
# J O S E
# 0 1 2 3 
# Siempre empieza desde el 0
# 0X00 J - 0X01 O - 0X02 S - 0X03 E
print(nombre[0])
print(nombre[1])
print(nombre[2])
print(nombre[3])
#corchetes para la posicion
primera_letra=nombre[0]
print(primera_letra)
#Ultima letra de la cadena -1 o contar manualmente
ultima_letra=nombre[-2]
print(ultima_letra)
#caracteres de la posicion 1 hasta el -1
segunda_hastaelfinal=nombre[1:-1]
print(segunda_hastaelfinal)

segunda_hastaelfinal=nombre[1:]
print(segunda_hastaelfinal)

segunda_hastaelfinal=nombre[0:4]
print(segunda_hastaelfinal)
```

```python
#PIDE EL PRECIO DE LA CUENTA
cuenta=float(input("Ingresa tu numero de cuenta:"))
#PROPINA DEL 10%
propina=float(input("ingresar porcentaje de la propina"))
personas=int(input("Ingresa numero de personas:"))
propina2=propina*1/100
#Trabajar en float porque hay decimales
monto_propina=float(cuenta)*float(propina2)
Monto_a_pagar=float(cuenta)+float(monto_propina)
print("el monto a pagar es:",Monto_a_pagar)
print("El monto por cada persona es:",Monto_a_pagar/personas)
```
