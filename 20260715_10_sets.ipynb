```python
conjunto_a={1,3,5,7,9,9,1,3}
conjunto_b={3,5,7,12}
conjunto_c=[1,3,5,7,9,1,3]
print(conjunto_a)
```

```python
conjunto_c[0]=50
```

```python
print(conjunto_b)
print(conjunto_c)
```

```python
#Union dde conjutnos
conjunto_a.union(conjunto_b)
```

```python
#otra manera de unir
conjunto_a | conjunto_b
```

```python
#interseccion
conjunto_a.intersection(conjunto_b)
```

```python
#otra manera de intersectar dos conjuntos operador logico 
conjunto_a & conjunto_b
```

```python
#Diferencia
conjunto_a.difference(conjunto_b)
# A - B
```

```python
#otra manera de diferencia dos conjuntos operador logico 
conjunto_a - conjunto_b
```

```python
conjunto_b.difference(conjunto_a)
# B - A
```

```python
#Diferencia simetrica 
conjunto_a.symmetric_difference(conjunto_b)
```

```python
#Diferencia simetrica operador logico
conjunto_a ^ conjunto_b
```
