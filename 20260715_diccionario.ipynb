```python
mi_diccionario={
    "PERU":"SOL",
    "JAPON":"YEN",
    "CANADA":"DOLAR CANADIENSE",
    "USA":"DOLAR AMERICANO",
}
print(mi_diccionario)
```

```python
mi_diccionario["mexico"]="peso mexicano"
print(mi_diccionario)
```

```python
mi_diccionario["PERU"]="nuevo sol"
print(mi_diccionario)
```

```python
# EL GET TE ENTREGA UN VALOR INCLUSO SI ESTE NO ESTA EN EL DICCIONARIO
mi_diccionario.get("argentina","buenos aires")
```

```python
#Update sirve para reemplazar o actualizar(agregar) un valor nuevo
mi_diccionario.update({"china":"yuan"})  
print(mi_diccionario)
```

```python
mi_diccionario.setdefault("bolivia","pesos boliviano")
print(mi_diccionario)
```

```python
#Elimianr valores especificos de la llave
mi_diccionario.pop("PERU")
print(mi_diccionario)
```

```python
mi_diccionario1={
    "peru":"lima",
    "argentina":"buenos aires",
    "noruega":"oslo",
    "japon":"tokio"
}
print(len(mi_diccionario1))
```

```python
for key in mi_diccionario1:
    print(key)
    print(mi_diccionario1[key])
    print("="*10)
```

```python
for key,value in mi_diccionario1.items():
    print(key)
    print(value)
    print("="*10)
```
