## 1. Estudiantes inscritos en dos cursos 
estudiantes_python = { "Ana", "Luis", "María", "Carlos", "Sofía", "Diego" } 
estudiantes_javascript = { "María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria" } 

Encuentra: 
- Los estudiantes inscritos en ambos cursos. 
- Los estudiantes que solo están en Python. 
- Los estudiantes que solo están en JavaScript. 
- Todos los estudiantes sin repetir nombres. 

```python
estudiantes_python = { "Ana", "Luis", "María", "Carlos", "Sofía", "Diego" } 
estudiantes_javascript = { "María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria" } 
print("Estudiante de ambos cursos",estudiantes_javascript.intersection(estudiantes_python))
print("Estudiantes que solo estudian python",estudiantes_python.difference(estudiantes_javascript))
print("Estudiantes que solo estudian javascrypt",estudiantes_javascript.difference(estudiantes_python))
print("Total de estudiantes",estudiantes_javascript.union(estudiantes_python))
```

## 2. Comparar inventarios de dos almacenes 
almacen_arequipa = { 
"laptop", "teclado", "mouse", "monitor", "impresora", "audifonos" 
} 
almacen_lima = { 
"laptop", "mouse", "monitor", "tablet", "webcam", "parlantes" 
} 

Determina: 
- Los productos disponibles en ambos almacenes. 
- Los productos exclusivos del almacén de Arequipa. 
- Los productos exclusivos del almacén de Lima. 
- Todos los productos disponibles sin repetir. 
- Si ambos almacenes tienen exactamente el mismo inventario. 

```python
almacen_arequipa = { "laptop", "teclado", "mouse", "monitor", "impresora", "audifonos" } 
almacen_lima = { "laptop", "mouse", "monitor", "tablet", "webcam", "parlantes" } 
print("Los productos disponibles en ambos almacenes",almacen_arequipa.intersection(almacen_lima))
print("Los productos exclusivos del almacén de Arequipa",almacen_arequipa-almacen_lima)
print("Los productos exclusivos del almacén de Lima",almacen_lima-almacen_arequipa)
print("Todos los productos disponibles sin repetir",almacen_arequipa.union(almacen_lima))
print("Si ambos almacenes tienen exactamente el mismo inventario",almacen_arequipa==almacen_lima)
```

## 3. Validar permisos de un usuario 
permisos_usuario = { 
"leer", "editar", "comentar", "descargar" 
} 
permisos_requeridos = { 
"leer", "editar", "eliminar", "publicar" 
} 
Crea un programa que muestre: 
- Si el usuario posee todos los permisos requeridos. 
- Los permisos que le faltan. 
- Los permisos adicionales que tiene. 
- La cantidad total de permisos diferentes. 

```python
permisos_usuario = { "leer", "editar", "comentar", "descargar" } 
permisos_requeridos = { "leer", "editar", "eliminar", "publicar" }
print(" Si el usuario posee todos los permisos requeridos",permisos_usuario.issubset(permisos_requeridos)) 
print("Los permisos que le faltan",permisos_requeridos-permisos_usuario)
print("Los permisos adicionales que tiene",permisos_usuario-permisos_requeridos)
print("La cantidad total de permisos diferentes",len(permisos_usuario.symmetric_difference(permisos_requeridos)))
```

## 4. Números presentes en varias listas 
numeros_a = {2, 4, 6, 8, 10, 12, 14} 
numeros_b = {4, 6, 9, 10, 13, 14, 16} 
numeros_c = {1, 4, 6, 10, 11, 14, 18} 
Encuentra: 
- Los números presentes en los tres conjuntos. 

- Los números presentes en numeros_a y numeros_b, pero no en numeros_c. 
- Los números que solo pertenecen a numeros_a. 
- Todos los números diferentes.

```python
numeros_a = {2, 4, 6, 8, 10, 12, 14} 
numeros_b = {4, 6, 9, 10, 13, 14, 16} 
numeros_c = {1, 4, 6, 10, 11, 14, 18} 
print("Los números presentes en los tres conjuntos",numeros_a.intersection(numeros_b,numeros_c))
print("Los números presentes en numeros_a y numeros_b, pero no en numeros_c",(numeros_a.intersection(numeros_b))-numeros_c)
print("Los números que solo pertenecen a numeros_a",numeros_a-numeros_c-numeros_b)
print("Todos los números diferentes",(numeros_b-numeros_a-numeros_c)|(numeros_a-numeros_b-numeros_c)|(numeros_c-numeros_a-numeros_b))

```

## 5. Comparar habilidades de candidatos 
habilidades_candidato_1 = { "Python", "Git", "SQL", "Docker", "React" } 
habilidades_candidato_2 = { "Python", "JavaScript", "React", "Node.js", "Git" } 
habilidades_requeridas = { "Python", "Git", "SQL", "Docker" } 
Determina: 
- Las habilidades que comparten ambos candidatos. 
- Las habilidades exclusivas de cada candidato. 
- Qué habilidades requeridas le faltan a cada candidato. 
- Qué candidato cumple con más requisitos. 
- Todas las habilidades disponibles sin repetir. 

```python
habilidades_candidato_1 = { "Python", "Git", "SQL", "Docker", "React" } 
habilidades_candidato_2 = { "Python", "JavaScript", "React", "Node.js", "Git" } 
habilidades_requeridas = { "Python", "Git", "SQL", "Docker" } 
print("Las habilidades que comparten ambos candidatos",habilidades_candidato_1.intersection(habilidades_candidato_2))
print("Las habilidades exclusivas del candidato 1",habilidades_candidato_1.difference(habilidades_candidato_2))
print("Las habilidades exclusivas del candidato 2",habilidades_candidato_2.difference(habilidades_candidato_1))
print("Qué habilidades requeridas le falta a candidato 1",habilidades_requeridas.difference(habilidades_candidato_1))
print("Qué habilidades requeridas le falta a candidato 2",habilidades_requeridas.difference(habilidades_candidato_2))
if len(habilidades_requeridas.intersection(habilidades_candidato_1)) < len(habilidades_requeridas.intersection(habilidades_candidato_2)):
    print("El candidato electo es el 2")
else:
    print("El candidato electo es el 1")
print("Todas las habilidades disponibles sin repetir",habilidades_candidato_2|habilidades_candidato_1|habilidades_requeridas)
```
