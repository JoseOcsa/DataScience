#1. Eliminar duplicados 

Dada una lista con números repetidos, crea una nueva lista sin números repetidos. 
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4] 

```python
lista = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]

sin_repetir = []

for num in lista:
    if num not in sin_repetir:
        sin_repetir.append(num)

print(f"Lista que no tiene elementos repetidos:{sin_repetir}")
```

## 5. Promedio de notas 
Dada una lista de notas, calcula el promedio e indica si el estudiante aprobó o desaprobó. 
notas = [14, 18, 11, 16, 9, 13, 20]

```python
notas=[14,18,11,16,9,13,20]
suma_total=0
total_notas=len(notas)
for nota in notas:
    suma_total=suma_total+nota
promedio=suma_total/total_notas
print(f"El promedio es:{promedio}")

if promedio >= 12:
    print("Aprobo")
else:
    print("Desaprobo")
```

## 10. Invertir una lista sin usar reverse() 
Dada una lista de colores, invierte la lista sin usar el método .reverse(). 
colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado'] 

```python
colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado']
ultimo_indice=len(colores) - 1
print(f"Ultimo indice:{ultimo_indice}")
colores_invertidos=[]
while ultimo_indice >= 0:
    colores_invertidos.append(colores[ultimo_indice])
    ultimo_indice=ultimo_indice-1
print(colores_invertidos)
print(f"segunda forma:{colores[::-1]}")
```

## 11. Ordenar palabras por longitud 
Dada una lista de palabras, ordénalas desde la más corta hasta la más larga. 
palabras = ['casa', 'elefante', 'sol', 'computadora', 'mesa', 'programacion]

```python
palabras = ['casa', 'elefante', 'sol', 'computadora', 'mesa', 'programacion']
longitudes_palabras=[
    len(palabra)
    for palabra in palabras 
]
lista_de_listas=[]
for item in zip(palabras,longitudes_palabras):
    lista_de_listas.append(item)
sorted(lista_de_listas,key=lambda x:x[1])
```

## 13. Contar frecuencia de elementos 
Dada una lista de nombres repetidos, muestra cuántas veces aparece cada nombre. 
nombres = ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 'María', 'Carlos'] 

```python
nombres = ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 'María', 'Carlos']

revi = []

for nombre in nombres:
    if nombre not in revi:
        contador = 0

        for n in nombres:
            if nombre == n:
                contador += 1

        print(nombre, "aparece", contador, "veces")
        revi.append(nombre)
```

## 16. Dividir una lista en dos partes 
Dada una lista de números, divídela en dos mitades. 

numeros = [10, 20, 30, 40, 50, 60, 70, 80] 

```python
numeros5 = [10, 20, 30, 40, 50, 60, 70, 80] 
half_list=[]
last_list=[]
longitud_lista=len(numeros5)
indice_mitad=longitud_lista// 2
for idx,numero in enumerate(numeros):
    if idx < indice_mitad:
        half_list.append(numero)
    else:
        last_list.append(numero)

print(half_list)
print(last_list)
```

```python
#Usando Slicing
print(numeros5[:indice_mitad])
print(numeros5[indice_mitad:])
```

## 20. Crear una lista de estudiantes aprobados 
Dada una lista de estudiantes y otra lista con sus notas, crea una nueva lista solo con los nombres de los 
estudiantes que aprobaron. Considera que la nota mínima aprobatoria es 11. 
estudiantes = ['Ana', 'Luis', 'Carlos', 'María', 'Pedro'] 
notas = [18, 10, 14, 8, 16]

```python
estudiantes = ['Ana', 'Luis', 'Carlos', 'María', 'Pedro'] 
notas = [18, 10, 14, 8, 16] 
lista_aprobados=[]
for estudiante,nota in zip(estudiantes,notas):
    if nota > 11:
        lista_aprobados.append(estudiante)
print(f"Lista de aprobados:{lista_aprobados}")
```
