#1. Eliminar duplicados 

Dada una lista con números repetidos, crea una nueva lista sin números repetidos. 
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4] 

```python
lista = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]

sin_repetir = []

for num in lista:
    if num not in sin_repetir:
        sin_repetir.append(num)

print(sin_repetir)
```

## 12. Eliminar elementos vacíos 
Dada una lista con textos y strings vacíos, crea una nueva lista eliminando los elementos vacíos. 
textos = ['hola', '', 'python', '', 'listas', ' ', 'programar', '']

```python
textos = ['hola', '', 'python', '', 'listas', ' ', 'programar', '']
nueva_lista = []
for vacio in textos:
    if vacio != '' and vacio != ' ':
        nueva_lista.append(vacio)
print(nueva_lista)
```

## 13. Contar frecuencia de elementos 
Dada una lista de nombres repetidos, muestra cuántas veces aparece cada nombre. 
nombres = ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 'María', 'Carlos'] 

```python
nombres = ['Ana', 'Luis', 'Ana', 'Carlos', 'Luis', 'Ana', 'María', 'Carlos']

revi = []

for nombre in nombres:
    if nombre not in revi:
        contador = 0

        for n in nombres:
            if nombre == n:
                contador += 1

        print(nombre, "aparece", contador, "veces")
        revi.append(nombre)
```
