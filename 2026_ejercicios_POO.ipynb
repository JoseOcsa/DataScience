## Ejercicio 1. Clase Persona 
Crea una clase Persona con los atributos nombre y edad. 
- Define el constructor __init__ que reciba ambos valores. 
- Agrega un método saludar() que imprima: "Hola, soy X y tengo Y años". 
- Crea dos objetos distintos y llama al método en cada uno. 
Objetivo: entender __init__, self y la diferencia entre clase y objeto.

```python
class Persona:
    def __init__(self,nombre:str,edad:int)-> None:
        self.nombre=nombre
        self.edad=edad
    def saludar(self):
        print(f"Hola soy {self.nombre} y tengo {self.edad} años ")

#CREAR DOS OBJETOS
Jorge=Persona("Jorge",25)
Maria=Persona("Maria",20)
Luis=Persona("Luis",22)
# INVOCAR AL METODO DE LA CLASE MEDIANTE EL .SALUDAR()
Jorge.saludar()
Maria.saludar()
Luis.saludar()

```

Crea una clase Rectangulo con los atributos base y altura. 
- Implementa el método area() que devuelva base × altura. 
- Implementa el método perimetro() que devuelva 2 × (base + altura). 
- Agrega el método especial __str__ para que print(mi_rectangulo) muestre las medidas y los resultados.

```python
class Rectangulo:
    def __init__(self,base:float,altura:float)->None:
        self.base=base
        self.altura=altura
    def area(self)->float:
        return self.base*self.altura
    
    def perimetro(self)->float:
        return 2* (self.base+self.altura)
    def __str__(self)->None:
        return (f"Rectangulo de base {self.base} y altura {self.altura}")
    
    
        


mi_rectangulo=Rectangulo(3,5)
print(mi_rectangulo.area())
print(mi_rectangulo.perimetro())
#aparece la ubicacion en sexagesimal
print(mi_rectangulo)
```
