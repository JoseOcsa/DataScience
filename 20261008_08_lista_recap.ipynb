```python
import random
#Crear lista vacia
numeros_random=[]
#bucle para que sea 10 elementos
for _ in range(10):
#Generar nuevo elemento con randint e insertar con append    
    numeros_random.append(random.randint(0,20))

print(numeros_random)
```

```python
#False ascendente y true descendente
sorted(numeros_random,reverse=True)
```

```python
numeros_random.sort()
```

```python
numeros_random.sort(reverse=True)
print(numeros_random)
```

```python
lista_letras2=["r","t","y","u"]
lista_letras=["d","z","a"]
for letra in lista_letras2:
    lista_letras.append(letra)
print(lista_letras)
```

```python
lista_letras2=["r","t","y","u"]
lista_letras=["d","z","a"]
lista_letras.extend(lista_letras2)
print(lista_letras)
```
