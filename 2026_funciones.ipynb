```python
#de entero a booleano
def es_par(numero:int) -> bool:
    if numero % 2 ==0:
        return True
    return False
    
print(es_par(10))
print(es_par(5))

```

```python
def mayor(numero_1:int|float,numero_2:int|float)-> int|float:
    if numero_1 > numero_2:
        return numero_1
    return numero_2
print(mayor(12,10))
print(mayor(5,15))
```

```python
#funcion para el promedio de una lista que acepta flotantes y enteros , devuelve enteros o flotantes
def promedio(lista:list[int|float])->float:
    return sum(lista)/len(lista)
mi_lista=[1,2,3,4,5]
print(promedio(mi_lista))
```

```python
def contar_vocales(texto:str)->int:
    vocales = ["a","e","i","o","u"]
    contador =0 
    for caracter in texto:
        if caracter in vocales:
            contador += 1
    return contador
contar_vocales("la casa es verde")
```
